#Gráfico de Barras Vertical

Para este gráfico usaremos a função *.bar()*

```python
# Um gráfico com a renda média por faixa etária será nosso Hello World, vamos definir alguns valores fictícios
faixaEtaria = ['18-25', '26-35', '36-45', '46-55', '55+']
renda = [1805.45, 2458.12, 3752.15, 4120.89, 3486.22]

plt.figure(figsize = (8,5)) 
plt.bar(faixaEtaria, renda, color="blue")

# Aqui definimos as legendas de cada barra no eixo X
plt.xticks(faixaEtaria)

# A label para o eixo Y
plt.ylabel('Renda Média')

# A label para o eixo X
plt.xlabel('Faixa Etária (anos)')

# O título do gráfico
plt.title('Renda Média vs Faixa Etária')
 
# Chamamos o método show() para mostrar o gráfico na tela
plt.show()
```

#Gráfico de Barras Horizontal

Para este gráfico usaremos a função *.barh()*

```python
# Este gráfico simula a participação dos 5 produtos agrícolas mais produzidos no Brasil

# Definindo os produtos, que vão para o eixo X
produtos = ['Soja em grão', 'Cana de Açucar', 'Milho em grão', 'Café em grão', 'Mandioca']

# Definindo as participações que formarão o eixo Y
quantidade_pct = [33.6, 16.8, 10.3, 6.2, 3.8]

# Definindo a margem de erro
margem_erro = [3,2,1,2,2]

# Agora com o parametro xerr, que representa o erro  usando barh()
plt.barh(produtos, quantidade_pct, color='green', xerr=margem_erro)

# A partir daqui não muda nada
plt.ylabel("Produtos agrícolas")
plt.xlabel("Participação do produto (em %)")
plt.title("5 Produtos com maiores participação da indústria agrícola")

plt.show()
```

#Gráfico de Barras Agrupadas

Para este gráfico usaremos a função *.bar()*

```python
# Gráfico sobre notas de 3 alunos nas provas do semestre
notas_pedro = [8, 9, 7, 8]
notas_maria = [5, 10, 6, 9]
notas_jose  = [7, 7, 5, 8]
 

# Definindo a largura das barras
barWidth = 0.25

# Aumentando o gráfico
plt.figure(figsize=(10,5))

# Definindo a posição das barras
r1 = np.arange(len(notas_pedro))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Criando as barras
plt.bar(r1, notas_pedro, color='#6A5ACD', width=barWidth, label='Pedro')
plt.bar(r2, notas_maria, color='#6495ED', width=barWidth, label='Maria')
plt.bar(r3, notas_jose, color='#00BFFF', width=barWidth, label='José')
 
# Adiciando legendas as barras
plt.xlabel('Provas')
plt.xticks([r + barWidth for r in range(len(notas_pedro))], ['Prova 1', 'Prova 2', 'Prova 3', 'Prova 4'])
plt.ylabel('Notas')
plt.title('Representação das notas de 3 alunos em 4 provas do semestre')
 
# Criando a legenda e exibindo o gráfico
plt.legend()
plt.show()
```

# Gráfico de Barras Empilhadas:

Gráfico de barras onde estão representados vários valores em uma coluna só:

```python
# Neste gráfico vamos representar a quantidade de ouro, pedra e madeira de aldeias de 3 aldeias do Age Of Empires o_o
gold  = np.array((104, 50, 280))
stone = np.array((223, 140, 60))
wood  = np.array((95, 228, 65))

resources = ['Ouro','Pedra','Madeira']

villages = ['Village A','Village B','Village C']


# Aumentando o gráfico
plt.figure(figsize=(10,5))

# Criando as barras
plt.bar(villages, gold, color = 'gold')
plt.bar(villages, stone, color = 'grey', bottom = gold)
plt.bar(villages, wood, color = 'saddlebrown', bottom = gold + stone)

# Adiciando legendas as barras
plt.xlabel('Vilas')
plt.ylabel('Quantidade de Recursos')
plt.title('Gráfico de recursos de 3 vilas')
plt.legend(('Ouro', 'Pedra', 'Madeira'))

plt.show()
```

# Gráfico de Pie


```python
votos = np.array([842.201, 488.775, 553.424, 424.307, 272.500, 381.512, 261.386])
candi = ['Candidato A', 'Candidato B', 'Candidato C', 'Candidato D', 'Candidato E',
         'Candidato F', 'Candidato G']

cores=['gold', 'red', 'blue', 'magenta', 'green','lightskyblue', 'yellowgreen']
# o atributo explode indica que fatia do gráfico será destacada. No exemplo abaixo, será a primeira fatia. A quantidade de valores é igual ao número de fatias do gráfico. 
explode = (0.1, 0, 0, 0, 0, 0, 0)  # explode 1st slice

# Atribuindo um título ao gráfico
plt.title('Eleição 2020 - Total de Votos')

plt.pie(votos, explode=explode, labels=candi, colors=cores, autopct='%1.1f%%', shadow=True, startangle=90)

#Adiciona Legenda
plt.legend(candi, bbox_to_anchor=(1.3, 1.3),loc='upper right')

#Centraliza o gráfico
plt.axis('equal')

#Ajusta o espaçamento para evitar o recorte do rótulo
plt.tight_layout()

plt.show()
```

# Gráfico Boxplot

```python
import seaborn as sns

#carregando a base de dados contida na biblioteca seaborn
base_dados = sns.load_dataset('iris')
# Visualizando a base de dados
base_dados.head(10)
# Use essa flag caso esteja utilizando o jupyter notebook ou o console ipython
%matplotlib inline
```

```python
# Colocando os dados no formato para plotar os boxplot com matplotlib
setosa = []
versicolor = []
virginica = []

flores = ["setosa", "versicolor", "virginica"]
for j, i in enumerate(base_dados.iloc[:, 4]):
  if i == flores[0]:
    setosa.append(base_dados.iloc[j, 0])
  if i == flores[1]:
    versicolor.append(base_dados.iloc[j, 0])
  if i == flores[2]:
    virginica.append(base_dados.iloc[j, 0])
  dados_sepal_length = [setosa, versicolor, virginica]

# Tamanho do gráfico em polegadas
plt.figure(figsize =(11, 6))

#Plotando o boxplot das espécies em relação ao tamanho das sépalas
bplots = plt.boxplot(dados_sepal_length,  vert = 1, patch_artist = False)

colors = ['pink', 'lightblue', 'lightgreen']
c = 0
for i, bplot in enumerate(bplots['boxes']):
  bplot.set(color=colors[c], linewidth=3)
  c += 1

# Linhas de contorno
colorss = ['pink','pink', 'lightblue', 'lightblue', 'lightgreen', 'lightgreen']
c2 = 0
for whisker in bplots['whiskers']:
  whisker.set(color=colorss[c2], linewidth=3)
  c2+= 1

c3 = 0
for cap in bplots['caps']:
  cap.set(color=colorss[c3], linewidth=3)
  c3 +=1

c4 = 0
for median in bplots['medians']:
  median.set(color=colors[c4], linewidth=3)
  c4 +=1

# Adicionando Título ao gráfico
plt.title("Boxplot da base de dados Íris", loc="center", fontsize=18)
plt.xlabel("Tipos de Flores")
plt.ylabel("Comprimento das sépalas")

plt.show()
```

#Gráfico de Dispersão

Dados:

Para a variavel independente x, os valosres foram gerados usando a função random (100 valores)

Para a variavel dependente y, os valores foram gerados usando uma equação y = 3x + adicional (gerado por uma distribuição normal) 

```python
x = np.random.uniform(low = 0, high = 10, size = 100)
y = 3*x + np.random.normal(0, 2, len(x))
```

```python
plt.plot(x,y,'^')
```

Foi feito o ajuste dos dados. Foram calculados os coeficientes b1 (pendiente da reta) b0 (offset) 

```python
b1, b0 = np.polyfit(x,y,deg = 1)
print('valores do ajuste b1 = {} , b0 = {}'.format(round(b1,2),round(b0,2)) )
```

Melhorando o gráfico:

'.' para não ligar os pontos

alpha = 0.4 indica o grau de transparência


```python
plt.figure(figsize = (8,5)) 
plt.plot(x,y,'.',marker = '^', alpha = 0.4, label = 'Valores')
plt.plot(x, b0 + b1*x, color = 'red', alpha = 0.5, label = 'Ajuste')
plt.title('Regressão Linear')

plt.xlabel('Valores X')
plt.ylabel('Valores Y')
plt.legend()
plt.show()
```

#Gráfico de Distribuição Normal

```python
import scipy.stats as ss
```

```python
x = np.arange(-3, 7, .001)
y = ss.norm.pdf(x,0,1)
y0 = ss.norm.pdf(x,2,2)
```

```python
p1=ss.norm.cdf(1)-ss.norm.cdf(-1)
p2=ss.norm.cdf(2)-ss.norm.cdf(-2)
  
print('P(-1 < z < 1) = {:2.2f}%'.format(100*p1))
print('P(-2 < z < 2) = {:2.2f}%'.format(100*p2))
```

```python
cg = {'blue': '#4285F4', 'red': '#DB4437', 'yellow': '#F4B400' , 'green': '#0F9D58'}
plt.figure(figsize=(10,6))
plt.ylim(-0.15,0.45)
plt.plot(x,y,color = cg['red'], label = 'N(0,1)')
plt.plot(x, y0, color = cg['blue'], label = 'N(2,2)')
plt.legend()

x1= np.arange(-1,1,0.01)
y1= ss.norm.pdf(x1)

x2= np.arange(-2,2,0.01)
y2= ss.norm.pdf(x2)

plt.fill_between(x,y,color='red', alpha=0.1)
plt.fill_between(x1,y1,color='red', alpha=0.1)
plt.fill_between(x2,y2,color='red', alpha=0.1)

plt.plot([-1,-1],[-0.1,ss.norm.pdf(-1)],color='darkred',linestyle='dashed',linewidth=2)
plt.plot([1,1],[-0.1,ss.norm.pdf(1)],color='darkred',linestyle='dashed',linewidth=2)

plt.plot([-2,-2],[-0.15,ss.norm.pdf(-2)],color='darkred',linestyle='dashed',linewidth=2)
plt.plot([2,2],[-0.15,ss.norm.pdf(2)],color='darkred',linestyle='dashed',linewidth=2)

plt.plot([0,0],[-0.15,ss.norm.pdf(0)],color='darkred',linestyle='dashed',linewidth=1)

plt.arrow(2.3,-0.05,-1,0,head_width = 0.01, head_length = 0.3)
plt.arrow(-2.3,-0.05,1,0,head_width = 0.01, head_length = 0.3)

plt.arrow(3.3,-0.1,-1,0,head_width = 0.01, head_length = 0.3)
plt.arrow(-3.3,-0.1,1,0,head_width = 0.01, head_length = 0.3)

plt.annotate('N(0,1)', [0, 0.2], ha='center', va='center', size = 20)

plt.annotate('{:2.2f}%'.format(100*p1),[3.1,-0.05],ha='center',va='center',size=12)
plt.annotate('{:2.2f}%'.format(100*p2),[4.1,-0.1],ha='center',va='center',size=12)

pos=[-2,-1,0,1,2,3,4,5]
lab=['$\mu - 2\sigma$','$\mu=0$', '$\mu + 1\sigma$','$\mu + 2\sigma$','$\mu + 3\sigma$','$\mu + 4\sigma$','$\mu + 5\sigma$']

plt.xticks(pos, lab)
plt.annotate('Distribuição Normal Padrão', [-3.5, 0.49], va='top', size = 20, annotation_clip=False, weight='bold')
plt.show()
```
