# **Solução de uma equação de 2do grau**

Uma equação de segundo grau é dada por:
# $$ax^2 + bx + c = 0$$
onde $a, b, c$ são os coeficientes da equação. As soluções são representadas por:

# $$x_{1,2} = \frac{-b \pm \sqrt{b^2 - 4\cdot a\cdot c}}{2\cdot a}$$

#Método de Bhaskara
- Cálculo del determinante $\Delta \qquad \rightarrow \qquad \Delta = b^2 - 4\cdot a\cdot c$   
  $\Delta > 0 \qquad \rightarrow \qquad$  tem duas soluções reais diferentes $x_1$ e $x_2$ \\
  $\Delta = 0 \qquad \rightarrow \qquad$  tem duas soluções reais iguais $x$ \\
  $\Delta < 0 \qquad \rightarrow \qquad$  tem duas soluções imaginarias $x_1$ e $x_2$

```python
# definição da função que sera usada com 3 parâmetros de entrada
def Bhaskara(a, b, c):
    print('Resolvendo a equação usando Bhaskara:')
    delta = b**2 -4*a*c
    print('Discriminante = {}'.format(delta,2) )
    
    # verifica as posibilidades de delta 
    if delta<0:
        print("Por ter discriminante < 0 !!!! , essa equação não tem solução real")
        x = -b/(2*a) 
        xx = ((-delta)**0.5)/(2*a)
        print('Soluções imaginarias:')
        if x!=0:
            print("x1 = {} + {} i".format(round(x,2),round(xx,2))) 
            print('x2 = {} - {} i'.format(round(x,2),round(xx,2))) 
        else:
            print("x1 = {} i".format(round(xx,2))) 
            print('x2 = - {} i'.format(round(xx,2)))     
    elif delta==0:
        print("tem duas soluções iguais")  
        x12=-b/(2*a)
        print('x1,2 = {}'.format(round(x12,2))) 
    else:
        print("tem duas soluções diferentes")
        x1 = (-b + (delta)**0.5)/(2*a)
        x2 = (-b - (delta)**0.5)/(2*a)
        print('x1 = {}'.format(round(x1,2))) 
        print('x2 = {}'.format(round(x2,2))) 

```

```python
# presentação do programa
print("Solução de uma equação de 2do grau")    
print("**********************************\n") 

# os valores sao ingresados por teclado
print("Ingrese os valores dos coeficientes a,b,c")
a = input('valor do coeficiente quadrático   (a) = ')
b = input('valor do coeficiente linear       (b) = ')
c = input('valor do coeficiente independente (c) = ')

# cuidado os dados ingresados por teclado podem ser considerados como string
# melhor transformar para float
aa = float(a)
bb = float(b)
cc = float(c)

if aa==0:
      print("------------------> A função não é de 2do grau")
else:  
      print('\nResolver a equação => {} x^2 + {} x + {} = 0\n'.format(round(aa,2),round(bb,2),round(cc,2)))

      print("------------------> Sera usado o método de Bhaskara")
```
